```python
df = pd.DataFrame(data)
print(df)
```

```python
plt.scatter(df['Intrate'], df['Index'], color='red')
plt.title('Index Price vs Interest Rate', fontsize=14)
plt.xlabel('Interest Rate', fontsize=14)
plt.ylabel('Index Price', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['Unemprate'], df['Index'], color='red')
plt.title('Index Price vs Unemployment Rate', fontsize=14)
plt.xlabel('Unemployment Rate', fontsize=14)
plt.ylabel('Index Price', fontsize=14)
plt.grid(True)
plt.show()
```

```python
from sklearn import linear_model
import statsmodels.api as sm
```

```python
x = df[['Intrate','Unemprate']]
y = df['Index']

#with sklearn
regr = linear_model.LinearRegression()
regr.fit(x, y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
#with statsmodels
x = sm.add_constant(x) # adding a constant

model = sm.OLS(y, x).fit()
predictions = model.predict(x)

print_model = model.summary()
print(print_model)
```
